```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x,y=make_regression(n_samples=100,n_targets=1,n_features=1,n_informative=1,noise=20,random_state=13)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
```

```python
rr = Ridge(alpha=10)
rr.fit(x,y)
print(rr.coef_)
print(rr.intercept_)
```

```python
rr1 = Ridge(alpha = 100)
rr1.fit(x,y)
print(rr1.coef_)
print(rr1.intercept_)
```

```python
plt.plot(x,y,'b.')
plt.plot(x,lr.predict(x),'red',label='alpha = 0')
plt.plot(x,rr.predict(x),'yellow',label='alpha = 10')
plt.plot(x,rr1.predict(x),'green',label='alpha = 100')
plt.legend()
plt.show()
```

# create the Ridge class for 2d data 

```python
class MeraRidge:
    
    def __init__(self,alpha=0.1):
        self.alpha = alpha
        self.m = None
        self.b = None
    
    def fit(self,x,y):
        num = 0
        den = 0
        
        for i in range(x.shape[0]):
            num = num +(y[i]-y.mean()) * (x[i]-x.mean())
            den = den +(x[i]-x.mean()) * (x[i]-x.mean())
        
            self.m = num / (den + self.alpha)
            self.b = y.mean() - (self.m * x.mean())
        print(self.m , self.b)

    def predict(x):
        return self.m*x + self.b
    
```

```python
rr2 = MeraRidge(10)
```
