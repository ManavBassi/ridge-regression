```python
from sklearn.datasets import load_diabetes
import  numpy  as np
from sklearn.metrics import r2_score
```

```python
x,y=load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
reg = SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,learning_rate='constant',alpha=0.001)
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
print(r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

```python
from sklearn.linear_model import Ridge

regR = Ridge(alpha=0.001 ,max_iter = 500 ,solver='sparse_cg')
```

```python
regR.fit(x_train,y_train)
y_pred1=regR.predict(x_test)
print(r2_score(y_test,y_pred1))
print(regR.coef_)
print(regR.intercept_)
```

# Create Class using Mathematical Formulation

```python
class MeraRidge:

    def __init__(self,epochs,learning_rate,alpha):

        self.learning_rate = learning_rate
        self.epochs = epochs
        self.alpha = alpha
        self.coef_ = None
        self.intercept_ = None

    def fit(self,x_train,y_train):

        self.coef_ = np.ones(x_train.shape[1])
        self.intercept_ = 0
        beta = np.insert(self.coef_,0,self.intercept_)
        # adding one to x_train as first col
        x_train = np.insert(x_train,0,1,axis=1)

        for i in range(self.epochs):
            beta_der = np.dot(x_train.T,x_train).dot(beta) - np.dot(x_train.T,y_train) + self.alpha*beta
            beta = beta - self.learning_rate*beta_der

        self.coef_ = beta[1:]
        self.intercept_ = beta[0]
    def predict(self,x_test):
        return np.dot(x_test,self.coef_) + self.intercept_
        
```
